Create Lists for [Item, Quantity, Price] with 10 random values of your choice.

Prepare a CSV file with combinations of these values with the help of random index & looping statements and other control structures we discussed in class.

```python
import random
import csv

# Create Lists for Items, Quantity, and Unit Prices
items = ["Item1", "Item2", "Item3", "Item4", "Item5", "Item6", "Item7", "Item8", "Item9", "Item10"]
quantity = [random.randint(1, 10) for _ in range(10)]
unit_prices = [round(random.uniform(1.0, 10.0), 2) for _ in range(10)]

# Create a CSV File
with open("Lab05.csv", mode="w", newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Item", "Quantity", "UnitPrice", "TotalCost"])

    for _ in range(100):
        index = random.randint(0, 9)
        item = items[index]
        qty = quantity[index]
        price = unit_prices[index]
        total_cost = round(qty * price, 2)
        writer.writerow([item, qty, price, total_cost])

# Calculate Total Cost for Each Row
with open("Lab05.csv", mode="r") as file:
    reader = csv.DictReader(file)
    total_cost_first5 = 0
    total_cost_last10 = 0
    unique_items = {}

    for row in reader:
        total_cost = float(row["TotalCost"])
        if reader.line_num <= 5:
            total_cost_first5 += total_cost
        elif reader.line_num > 90:
            total_cost_last10 += total_cost

        item = row["Item"]
        qty = int(row["Quantity"])
        if item in unique_items:
            unique_items[item]["Total Quantity"] += qty
            unique_items[item]["Total Cost"] += total_cost
        else:
            unique_items[item] = {"Total Quantity": qty, "Total Cost": total_cost}

# Print Results
print("Total Cost for the first 5 rows:", total_cost_first5)
print("Total Cost for the last 10 rows:", total_cost_last10)
print("Total Cost for Unique Items:")
for item, data in unique_items.items():
    print(f"{item} | {data['Total Quantity']} | {data['Total Cost']}")

# Calculate and Print Min & Max Price for Each Item
for item in items:
    item_prices = [float(row["UnitPrice"]) for row in reader if row["Item"] == item]
    if item_prices:
        min_price = min(item_prices)
        max_price = max(item_prices)
        print(f"{item} - Min Price: {min_price} | Max Price: {max_price}")
```
