```python
def manual_calculation(data): 
    min_val = data[0]
    max_val = data[0]
    total = 0
    count = 0
    median = 0

    for num in data:
        total += num
        count += 1

        if num < min_val:
            min_val = num
        elif num > max_val:
            max_val = num

    for k in range(0, count, 1):  # median calculation
        for j in range(k + 1, count):
            if data[k] > data[j]:
                temp = data[k]
                data[k] = data[j]
                data[j] = temp

    if count % 2 == 0:
        val1 = int(count / 2)
        val2 = val1 + 1
        median = (data[val1] + data[val2]) / 2

    check = 0
    temp = 2
    mode = 0

    for num in data:
        check = 0
        for j in data:
            if num == j:
                check += 1
        if check >= temp:
            temp = check
            mode = num

    print("Manually Calculated Values:", "\nMean is=", (total / count), "\nMin is =", min_val, "\nMax is=", max_val,
          "\nRange is =", (max_val - min_val), "\nMedian is=", median)
    print("Mode is= ", mode)


def predefined_calculation(data):  
    print("\nPredefined Calculated Values:", "\nMean is=", sum(data) / len(data), "\nMin is=", min(data),
          "\nMax is=", max(data), "\nRange is=", max(data) - min(data))


data = [761, 123, 585, 275, 194, 840, 999, 622, 763, 40, 825, 687, 402, 338, 611, 987, 739, 95, 819, 568, 320, 126, 688, 948, 514, 223, 751, 379, 857, 767, 492, 162, 437, 934, 430, 343, 725, 511, 580, 678, 708, 823, 346, 179, 786, 663, 195, 890, 267, 502, 214, 300, 786, 961, 235, 559, 992, 919, 948, 318, 57, 787, 653, 317, 305, 172, 966, 182, 970, 17, 89, 624, 629, 595, 485, 752, 984, 978, 460, 124, 321, 256, 614, 852, 811, 561, 127, 207, 406, 615, 888, 391, 691, 470, 911, 401, 137, 600, 478, 682]
manual_calculation(data)
predefined_calculation(data)
print("Since both the values are similar, we can use either manual calculation or pre-defined calculation function")
```
